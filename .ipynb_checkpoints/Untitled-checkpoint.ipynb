```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url='https://dondeverlo.com/futbol/equipo/athletic'
res = requests.get(url)
html = res.text
soup = BeautifulSoup(html,'html.parser')
```

```python
print(soup.prettify())
```

```python
##Partido
partidos1 = soup.findAll('h2',{"class":'Event__title'})
partido = partidos1[0].find("a").text

##partidos2 = [((e.find_all("a")[0]).get('href')) for e in partidos1]
##partidos2 = [e.get('h2') for e in partidos1]


print(partidos1)
print(partido)
```

```python
#Fecha
date = soup.findAll('div',{"class":'Event__when'})
date2 = date[0].text

##partidos2 = [((e.find_all("a")[0]).get('href')) for e in partidos1]
##partidos2 = [e.get('h2') for e in partidos1]


print(date)
print(date2)
```
